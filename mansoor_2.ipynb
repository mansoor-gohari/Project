```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### داده‌های قد و وزن

```python
data = pd.read_csv("weight-height.csv")
```

### جدا کردن ویژگی‌ها  و برچسب‌ها 

```python
x = data["Weight"].values.reshape(-1, 1)  
y = data["Height"].values.reshape(-1, 1)  
```

### تقسیم داده‌ها به دو بخش آموزشی و آزمایشی

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

#### ایجاد مدل رگرسیون خطی

```python
model = LinearRegression()
```

### آموزش مدل با داده‌های آموزشی

```python
model.fit(x_train, y_train)
```

### پیش‌بینی با داده‌های آزمایشی

```python
y_pred = model.predict(x_test)
```

### رسم نمودار رگرسیون

```python
plt.scatter(x_test, y_test, color='blue', label='real')
plt.plot(x_test, y_pred, color='red', linewidth=2, label='guess')
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Regression")
plt.legend()
plt.show()
```
